```python
import pandas as pd  # import libraries
import seaborn as sns  #for graphs
import matplotlib.pyplot as plt  # for plot
import numpy as np
```

```python
dataset=pd.read_csv("/content/Stars_Names (1).csv") # Import data that has undergone data cleaning and data preprocessing
```

```python
dataset
```

```python
# Convert to datetime format
dataset['Approval Date'] = pd.to_datetime(dataset['Approval Date'])
```

```python
# Sample date column
df = pd.DataFrame(dataset)
```

```python
df
```

```python
# Filter only zodiac constellations (12 classical signs)

zodiac_signs = ['Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo',
                'Libra', 'Scorpius', 'Sagittarius', 'Capricornus', 'Aquarius', 'Pisces']

zodiac_df = df[df['Constellation'].isin(zodiac_signs)]  ##FILTER## It contains stars with mentioned zodfaic signs only

# Count stars per zodiac sign
sign_counts = zodiac_df['Constellation'].value_counts().reindex(zodiac_signs, fill_value=0)

plt.figure(figsize=(14, 6)) # Size of figure

# Bar Chart
plt.subplot(1, 2, 1)
bars = plt.bar(sign_counts.index, sign_counts.values, color=plt.cm.tab20.colors)
plt.title('Zodiac Star Distribution (Bar Chart)', fontsize=14)
plt.xlabel('Zodiac Sign', fontsize=12)
plt.ylabel('Number of Stars', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y', linestyle=':', alpha=0.7)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom')

# Pie Chart
plt.subplot(1, 2, 2)
plt.pie(sign_counts, labels=sign_counts.index, autopct='%1.1f%%',
        colors=plt.cm.tab20.colors, startangle=90)
plt.title('Zodiac Star Distribution (Pie Chart)', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
#medium Q2
plt.figure(figsize=(10, 6))
sorted_months = sorted(df['Approval Month'].unique()) # Sort months

# Seaborn is used to create boxplot
sns.boxplot(x='Approval Month', y='Approval Year', data=df, order=sorted_months) # Reference the 'Month_Number' column within the dataframe
plt.title('Boxplot of Star Approvals: Year vs Month')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#hard Q2 :
# Using PCA to reduce two variables (Approval Year, Approval Month)to 2 principle componenets
# Selecting features for PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
features = ['Approval Year', 'Approval Month']
X = df[features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with the principal components
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

df_pca['Approval Month'] = df['Approval Month']
df_pca['Approval Year'] = df['Approval Year']

# Plot the PCA components
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', data=df_pca, hue='Approval Month', palette='tab10', s=100)
plt.title('PCA on Year and Month')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Approval Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#hard Q1
from sklearn.preprocessing import LabelEncoder, StandardScaler # Converts categorical to numeraical data that is encoding
from sklearn.cluster import KMeans # For clustrring

# Select categorical columns to encode and numerical columns for clustering
categorical_cols = ['Name', 'Constellation', 'Bayern Designation', 'Designation']
numerical_cols = ['Approval Year', 'Approval Month']

# Encoding categorical data to numerical column
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

clustering_features = df[categorical_cols + numerical_cols]

# Apply kmeans clustering
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(clustering_features)

# Visualizing the clusters using first two features
plt.scatter(df[categorical_cols[0]], df[categorical_cols[1]], c=df['Cluster'], cmap='viridis')
plt.xlabel(categorical_cols[0])
plt.ylabel(categorical_cols[1])
plt.title('KMeans Clustering of Stars')
plt.show()
print(df.head())
```

```python
#Medium Q3
# Scatter plot year vs month
plt.figure(figsize=(8, 6))
plt.scatter(df["Approval Year"], df["Approval Month"], color='purple')
plt.title("Scatter Plot of Approval Dates")
plt.xlabel("Approval Year")
plt.ylabel("Approval Month")
plt.yticks(range(1, 13))  # Months 1-12
plt.grid(True)
plt.show()
```

```python
#medium Q1
# Group by Year and Month
# To get number of stars noted in particular year and month
# Intensity of color represents no. of stars
heatmap_data = df.groupby(["Approval Year", "Approval Month"]).size().unstack(fill_value=0)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu", linewidths=0.5, linecolor='gray')
plt.title("Approvals Heatmap by Year and Month")
plt.xlabel("Approval Month")
plt.ylabel("Approval Year")
plt.show()
```

```python
#easy Q1
# Creating 'YearMonth' column for monthly grouping
# Changed: Using .to_period('M') to ensure it's datetime-like for time series plotting
df['YearMonth'] = df['Approval Date'].dt.to_period('M')

# Group by 'YearMonth' and count approvals
monthly_counts = df.groupby('YearMonth').size().reset_index(name='Approval Count')

monthly_counts['YearMonth'] = monthly_counts['YearMonth'].dt.to_timestamp() # no of approvals in particaular month and year

# Plotting
plt.figure(figsize=(14, 6))
plt.plot(monthly_counts['YearMonth'], monthly_counts['Approval Count'], marker='o', linestyle='-')
plt.title('Star Name Approval Density Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Number of Approvals')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Medium Q2
# Count approvals per day
daily_counts = df.groupby(df['Approval Date'].dt.date).size().reset_index(name='Approval Count')

# Plot KDE of approval counts
plt.figure(figsize=(10, 5))
sns.kdeplot(data=daily_counts, x='Approval Count', fill=True, bw_adjust=0.5)
plt.title('KDE Plot of Daily Star Approval Counts')
plt.xlabel('Number of Approvals per Day')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()

# Here x axis represent no of approvals per day and y axis represent no of days when these data of approvals was seen
# So approximately 8 approvals were found on most of the days
# And approximately 60 approvals were made on fewer days so this becomes anomaly
```

```python
#hard Q2 #anamoly detection
df = df.sort_values('Approval Date')

# Getting monthly approvals
monthly = df.groupby(df['Approval Date'].dt.to_period('M')).size()
monthly.index = monthly.index.to_timestamp()

# Setting thresholds for high and low anomalies (top 10% and bottom 20%)
high_threshold = monthly.quantile(0.9)
low_threshold = monthly.quantile(0.2)

# Identify anomalies
high_anomalies = monthly[monthly > high_threshold]
low_anomalies = monthly[monthly < low_threshold]

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(monthly.index, monthly, 'b-', label='Monthly Approvals', alpha=0.7)

# Highlighting high anomalies in red
plt.scatter(high_anomalies.index, high_anomalies,
            color='red', s=100, label='High Anomalies')
for date, count in high_anomalies.items():
    plt.text(date, count, f'{count} approvals\n{date.strftime("%b %Y")}',
             ha='center', va='bottom', fontsize=8)

# Highlighting low anomalies in green
plt.scatter(low_anomalies.index, low_anomalies,
            color='green', s=100, label='Low Anomalies')
for date, count in low_anomalies.items():
    plt.text(date, count, f'{count} approvals\n{date.strftime("%b %Y")}',
             ha='center', va='top', fontsize=8)

plt.title('Star Name Approvals with High and Low Anomalies', fontsize=14)
plt.ylabel('Number of Approvals', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print results
print("High Approval Anomalies (Top 10%):")
print(high_anomalies.sort_values(ascending=False))
print("\nLow Approval Anomalies (Bottom 10%):")
print(low_anomalies.sort_values())
```

```python
#hard Q1
# Detect outliers - dates before 2016 or after 2017
outliers = df[(df['Approval Year'] < 2016) | (df['Approval Year'] > 2017)]

# Creating the plot
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Plotting all data points in light gray
plt.scatter(df['Approval Date'], [1]*len(df), color='lightgray', alpha=0.5, label='Normal Approvals')

# Plotting outliers in red
plt.scatter(outliers['Approval Date'], [1]*len(outliers), color='red', label='Outlier Approvals')

plt.title('Star Name Approval Dates with Outliers Highlighted', fontsize=14)
plt.xlabel('Approval Date', fontsize=12)
plt.yticks([])
plt.legend()
plt.xticks(rotation=90)

# Highlighting the outlier clusters
plt.annotate('Early Outliers (2015)', xy=(pd.to_datetime('2015-12-15'), 1),
             xytext=(pd.to_datetime('2015-09-01'), 1.02),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Late Outliers (2018)', xy=(pd.to_datetime('2018-06-15'), 1),
             xytext=(pd.to_datetime('2018-03-01'), 1.02),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()

outliers = df[(df['Approval Year'] < 2016) | (df['Approval Year'] > 2017)]

# Getting first 10 and last 10 outliers
first_10_outliers = outliers.head(10)
last_10_outliers = outliers.tail(10)

print("First 10 outlier stars:")
print(first_10_outliers[['Name', 'Constellation', 'Approval Date']].to_string(index=False))

print("\nLast 10 outlier stars:")
print(last_10_outliers[['Name', 'Constellation', 'Approval Date']].to_string(index=False))
```
